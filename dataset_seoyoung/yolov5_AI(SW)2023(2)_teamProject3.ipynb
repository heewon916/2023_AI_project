# 2. 모델 학습(Train)
- batch: 한 번에 처리하는 데이터의 샘플 수(큰 배치 크기는 더 안정적인 그래디언트 계산을 제공하지만 메모리를 많이 차지하고 훈련이 느려질 수 있다. 반면, 작은 배치 크기는 더 빠른 훈련을 제공할 수 있지만 노이즈가 높을 수 있다.)
- epochs: 전체 데이터셋을 몇 번 반복하여 모델을 학습시킬지 나타내는 하이퍼파라미터
- weights: 모델의 가중치를 나타낸다. (yolov5.pt)
